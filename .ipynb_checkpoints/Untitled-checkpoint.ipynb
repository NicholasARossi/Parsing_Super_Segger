```python
def num_bins_calculator(points):
    return np.sqrt(points/5)

def calc_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    mi = mutual_info_score(None, None, contingency=c_xy)
    return mi

def mutual_info_score(labels_true, labels_pred, contingency=None):

    if contingency is None:
        labels_true, labels_pred = check_clusterings(labels_true, labels_pred)
        contingency = contingency_matrix(labels_true, labels_pred, sparse=True)

    if isinstance(contingency, np.ndarray):
        # For an array
        nzx, nzy = np.nonzero(contingency)
        nz_val = contingency[nzx, nzy]
    elif sp.issparse(contingency):
        # For a sparse matrix
        nzx, nzy, nz_val = sp.find(contingency)
    else:
        raise ValueError("Unsupported type for 'contingency': %s" %
                         type(contingency))

    contingency_sum = contingency.sum()
    pi = np.ravel(contingency.sum(axis=1))
    pj = np.ravel(contingency.sum(axis=0))
    log_contingency_nm = np.log(nz_val)
    contingency_nm = nz_val / contingency_sum
    # Don't need to calculate the full outer product, just for non-zeroes
    outer = pi.take(nzx) * pj.take(nzy)
    log_outer = -np.log(outer) + np.log(pi.sum()) + np.log(pj.sum())
    mi = (contingency_nm * (log_contingency_nm - np.log(contingency_sum)) +
          contingency_nm * log_outer)
    return mi.sum()
```

```python
from os import listdir
names=listdir('MGC_clists')
del names[0]
fig6,ax6=plt.subplots()
j=0
for name in names[0:3]:
    mat_contents = sio.loadmat('MGC_clists/'+name)
    data_frame=mat_contents['data3D']
    cell_numbers,featurs,time_range=data_frame.shape
    variance=[]
    times=[]
    for time in np.arange(time_range):
        
        indxs=[~np.isnan(data_frame[:,5,time].ravel())]
        try:
            bins=num_bins_calculator(len(data_frame[:,7,time][indxs]))
            variance.append(calc_MI(data_frame[:,7,time][indxs],data_frame[:,9,time][indxs], bins))
            times.append(time)
        except:
            print('no points')
#         variance.append(np.corrcoef(data_frame[:,7,time][indxs],data_frame[:,9,time][indxs])[0][1])

#         variance.append(np.nanstd(data_frame[:,9,time])/np.nanmean(data_frame[:,9,time]))
    plt.plot(times,variance,linewidth=3)
    j+=1
fig6
```
